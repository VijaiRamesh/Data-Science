## Q10) Find the list of Affordable Restaurants?
### The criteria for Affordable Restaurants would be:-
### 1) Low Price 2) High Rated
### First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
### Let me explain:-The most expensive restaurant has an average meal cost= 2000. We'll try to stay economical and only pick the
### restaurants that are 1/4th of 2000.

```python
max_price = dataset['Average Cost for two'].max()
one_fourth_price = max_price/4
one_fourth_price
```

```python
max_price
```

```python
dataset.head()
```

### Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restaurant

```python
aff_rest_df = dataset[['Restaurant Name', 'Average Cost for two', 'Cuisines', 'Locality', 'Price range']]
aff_rest_df = aff_rest_df[aff_rest_df['Average Cost for two'] <= 500]
aff_rest_df.sort_values(by='Average Cost for two', inplace=True)
```

```python
aff_rest_df
```

# Finding the highest rated list of restaurants

```python
highrate_rest_df = dataset[['Restaurant Name', 'Average Cost for two', 'Cuisines', 'Locality', 'Price range','Aggregate rating']]
highrate_rest_df = highrate_rest_df[highrate_rest_df['Aggregate rating'] >= 4.5]
highrate_rest_df.sort_values(by='Average Cost for two', inplace=True)
```

```python
highrate_rest_df
```

#### Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !!

```python
highrate_aff_df = pd.merge(aff_rest_df, highrate_rest_df, how='inner', on=['Restaurant Name', 'Locality'])
highrate_aff_df = highrate_aff_df[['Restaurant Name', 'Average Cost for two_x', 'Cuisines_x', 'Locality', 'Price range_x']]
highrate_aff_df.rename(columns={'NAME':'Restaurant Name', 'Average Cost for two_x':'Average Cost for two', 'Cuisines_x':'Cuisines',
 'Locality':'Locality', 'Price range_x':'Price range'},inplace=True)
```

```python
highrate_aff_df
```

## Q10) Find the list of most Reliable Restaurants?
### The criteria for most Reliable Restaurants would be:-
### 1) Low Price 2) High Rated 3) Large No. of Votes
### First step will be to find the restaurants with Votes greater than Mean of Votes

```python
mean_votes = dataset['Votes'].mean()
```

```python
mean_votes
```

### Finding list of restaurants that have Votes greater than and equal to Mean of Vote

```python
mean_rest_df = dataset[['Restaurant Name', 'Average Cost for two', 'Cuisines', 'Locality', 'Price range', 'Votes']]
mean_rest_df = mean_rest_df[mean_rest_df['Votes'] > 1384]
mean_rest_df.sort_values(by='Votes', inplace=True)
```

```python
mean_rest_df
```

## These are the most reliable, highest rated and affordable restaurants:-
#### We obtain this dataframe by simply taking the intersection of highrate_aff_df & mean_rest_df
#### This dataframe obtained below shows the restaurants whose:
#### Cost is below 500
#### Rating is above 4.5
#### Votes are above 1384

```python
reliable_rest_df = pd.merge(mean_rest_df, highrate_aff_df, how='inner', on=['Restaurant Name', 'Locality'])
reliable_rest_df = reliable_rest_df[['Restaurant Name', 'Average Cost for two_x', 'Cuisines_x', 'Locality', 'Price range_x']]
reliable_rest_df.rename(columns={'Restaurant Name':'Restaurant Name', 'PRICE_x':'PRICE', 'CUSINE_CATEGORY_x':'CUSINE_CATEGORY',
 'Locality':'Locality', 'Price range_x':'Price range'},inplace=True)
```
